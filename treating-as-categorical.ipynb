```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('cces_example_data.csv')
```

```python
data['pid_cat'] = data['pid3'].astype('category')
data['pid_cat'].value_counts()
```

```python
data['pid_lab'] = data['pid_cat'].cat.rename_categories({1: "Democrat", 2: "Republican", 3: "Independent/Other"})
data['pid_lab'].value_counts()
```

```python
#Creating a figure with two axes
fig2, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
pid = data['pid_lab'].value_counts()
#Setting titles for axises
ax1.set_title('Age Histogram')
ax2.set_title('Political Identity')
#creating histograms for the age and political identity distibution (im not sure whether the pid3 stands for the political identity but thats what google told me )
ax1.hist(x= data['age'], color='black')
ax2.bar(x = pid.index, height = pid, color='purple')
plt.show()
```
