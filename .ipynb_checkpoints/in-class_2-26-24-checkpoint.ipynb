#### Group 1 Demo

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
dia = pd.read_csv('Diabetes.csv')
dia.head()
```

More info at [Controlling figure aesthetics](https://seaborn.pydata.org/tutorial/aesthetics.html)
and [seaborn.set_theme](https://seaborn.pydata.org/generated/seaborn.set_theme.html)

Valid options are: 'white', 'whitegrid', 'dark', 'ticks'

```python
sn.set_theme(style='darkgrid')
```

```python
dia['Garbage naME_sd'] = dia['BMI']
```

```python
sn.scatterplot(x=dia['BMI'], y=dia['Glucose'])
plt.show()
```

```python
sn.scatterplot(x=dia['Garbage naME_sd'], y=dia['Glucose'])
plt.show()
```

Information on the [relplot from the documentation](https://seaborn.pydata.org/generated/seaborn.relplot.html)

```python
sn.relplot(data=dia, x='BMI', y='Glucose',hue='Pregnancies', alpha = .5)
plt.show()
```

```python
sn.relplot(data=dia, x='BMI', y='Glucose',hue='Pregnancies', size = 'Pregnancies')
plt.show()
```

```python
sn.relplot(data=dia, x='BMI', y='Glucose',hue='Pregnancies', size = 'Pregnancies', sizes = (1,100))
plt.show()
```

```python
sn.relplot(data=dia, x='BMI', y='Glucose',hue='Pregnancies', size = 'Pregnancies', sizes = (1,100))
plt.xlim(18,35)
plt.ylim(60, 140)
plt.show()
```

#### Group 2 Demo/Challenge

```python
sd = pd.read_csv('Salary Data.csv')
sd.head()
```

```python
sn.set(style='whitegrid')
```

```python
plt.figure()
sn.scatterplot(x = 'Age', y = 'Salary', data = sd)
plt.title('Default Scatter Plot: Ave vs Salary')
plt.show()
```

```python
plt.figure()
sn.scatterplot(x = 'Age',\
               y = 'Salary',\
               hue= 'Education Level',\
               style = 'Gender',\
               data = sd)
plt.title('Scatter Plot: Ave vs Salary by Education Level and Gender')
plt.show()
```

For a guide on using matplotlib (and seaborn) named colors, see [this cheatsheet](https://matplotlib.org/stable/gallery/color/named_colors.html).

```python
ed_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
plt.figure()
sn.scatterplot(x = 'Age',\
               y = 'Salary',\
               hue= 'Education Level',\
               style = 'Gender',\
               palette = ed_colors,\
               data = sd)
plt.title('Scatter Plot: Ave vs Salary by Education Level and Gender')
plt.show()
```

For all possible matplotlib (and seaborn) marker options, see the [marker documentation](https://matplotlib.org/stable/api/markers_api.html)

```python
ed_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gen_markers = {'Female': 'o', 'Male': 's'}
plt.figure()
sn.scatterplot(x = 'Age',\
               y = 'Salary',\
               hue= 'Education Level',\
               style = 'Gender',\
               palette = ed_colors,\
               markers = gen_markers,\
               data = sd)
plt.title('Scatter Plot: Ave vs Salary by Education Level and Gender')
plt.show()
```

```python
ed_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gen_markers = {'Female': 'o', 'Male': 's'}
plt.figure(figsize=(12,8))
sn.set_style('ticks')
sn.scatterplot(x = 'Age',\
               y = 'Salary',\
               hue= 'Education Level',\
               style = 'Gender',\
               palette = ed_colors,\
               markers = gen_markers,\
               data = sd)
plt.title('Scatter Plot: Ave vs Salary by Education Level and Gender')
plt.show()
```

All possible legend arguments and locations can be found [in the documentation for pyplot.legend](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html)

```python
ed_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gen_markers = {'Female': 'o', 'Male': 's'}
plt.figure(figsize=(12,8))
sn.set_style('ticks')
sn.scatterplot(x = 'Age',\
               y = 'Salary',\
               hue= 'Education Level',\
               style = 'Gender',\
               palette = ed_colors,\
               markers = gen_markers,\
               data = sd)

plt.legend(ncol=2, frameon=False, loc='lower right')
plt.title('Scatter Plot: Ave vs Salary by Education Level and Gender')
plt.show()
```

```python
dat_b = sd[sd['Education Level'] == "Bachelor's"]
gen_colors = {'Female': 'red', 'Male': 'blue'}
plt.figure(figsize=(10,8))
sn.set_style('ticks')
sn.scatterplot(x = 'Age',\
               y = 'Salary',\
               hue= 'Gender',\
               palette = gen_colors,\
               data = dat_b)
plt.xlim(25, 50)
plt.ylim(50000, 250000)
plt.legend(ncol=2, frameon=False, loc='upper left')
plt.title("Scatter Plot: Ave vs Salary on Gender Among Bachelor's Degree Holders")
plt.show()
```

```python
dat_m = sd[sd['Education Level'] == "Master's"]
gen_colors = {'Female': 'red', 'Male': 'blue'}
plt.figure(figsize=(10,8))
sn.set_style('ticks')
sn.scatterplot(x = 'Age',\
               y = 'Salary',\
               hue= 'Gender',\
               palette = gen_colors,\
               data = dat_m)
plt.xlim(25, 50)
plt.ylim(50000, 250000)
plt.legend(ncol=2, frameon=False, loc='upper left')
plt.title("Scatter Plot: Ave vs Salary on Gender Among Master's Degree Holders")
plt.show()
```

```python
dat_p = sd[sd['Education Level'] == "PhD"]
gen_colors = {'Female': 'red', 'Male': 'blue'}
plt.figure(figsize=(10,8))
sn.set_style('ticks')
sn.scatterplot(x = 'Age',\
               y = 'Salary',\
               hue= 'Gender',\
               palette = gen_colors,\
               data = dat_p)
plt.xlim(25, 50)
plt.ylim(50000, 250000)
plt.legend(ncol=2, frameon=False, loc='upper left')
plt.title("Scatter Plot: Ave vs Salary on Gender Among PhD Holders")
plt.show()
```

```python
sn.barplot(x='Years of Experience', y='Education Level', hue='Gender', data = sd)
plt.show()
```

The better figure for continuous data is a boxplot instead of a barplot.

```python
sn.boxplot(x='Years of Experience', y='Education Level', hue='Gender', data = sd)
plt.show()
```

```python
sn.barplot(x='Education Level', y='Years of Experience', hue='Gender', capsize=0.1,data = sd)
plt.show()
```

```python
sn.barplot(x='Education Level', y='Years of Experience', hue='Gender', capsize=0.1,data = sd, errorbar=None)
plt.show()
```

```python
sn.barplot(x='Education Level', y='Years of Experience', hue='Gender', palette = 'dark:blue',  hue_order = ['Female', 'Male'], data = sd, errorbar=None)
plt.show()
```

```python
sn.barplot(x='Education Level', y='Years of Experience', hue='Gender', palette = 'hls',  hue_order = ['Female', 'Male'], data = sd, errorbar=None)
plt.show()
```
