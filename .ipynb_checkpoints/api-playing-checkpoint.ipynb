```python
import requests
import pandas as pd
import json
from io import StringIO
```

```python
wiki_url = 'https://en.wikipedia.org/wiki/FIFA_World_Cup'
```

```python
wiki_worldcup_response = requests.get(wiki_url)
```

```python
wiki_worldcup_response.status_code
```

```python
wiki_worldcup_response.text[0:20]
```

### Remember to install lxml if first use

```python
worldcup_tables = pd.read_html(StringIO(wiki_worldcup_response.text))
len(worldcup_tables)
```

Using The [Aviation Weather](https://aviationweather.gov/data/api/#/Data/dataAirport) endpoints.

```python
airports = requests.get('https://aviationweather.gov/api/data/airport?ids=@WA&format=json')
```

```python
airports.status_code
```

```python
airports_list = json.loads(airports.text)
```

```python
airports_df = pd.DataFrame(airports_list)
airports_df.shape
```
