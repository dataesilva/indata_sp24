```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
reddit = pd.read_csv('reddit_data_example.csv')
reddit.shape
```

```python
reddit['datetime'] = pd.to_datetime(reddit['Date'])
reddit_cd = reddit.groupby(reddit['datetime'].dt.date)['Reddit Score'].count()
date_range = pd.date_range(start = '2017-06-20', end = '2023-02-09', freq='D')
rds = pd.Series(reddit_cd, index = date_range)
```

```python
viz = pd.DataFrame({'count':rds, 'confidence':['observed']*len(rds)}, index = date_range)
viz.head()
```

```python
test = ['observed' if x != np.nan else 'expected' for x in viz['count']]
```
