```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
group1 = np.random.normal(0, 1, 20)
```

```python
fig1, ax1 = plt.subplots()
group1 = np.random.normal(0, 1, 20)
group2 = np.random.normal(1, .5, 20)
ax1.boxplot(x = group1, positions = [1], patch_artist = True,
           boxprops = dict(facecolor = "lightblue"))
ax1.boxplot(x = group2, positions = [2],
           patch_artist = True,
           boxprops = dict(facecolor = "pink"))

plt.show()
```
