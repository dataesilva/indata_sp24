```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
commute = pd.read_csv('commute.csv')
commute.head()
```

```python
fig1, ax1 = plt.subplots()
ax1.scatter(x = commute['TripID'], y = commute['CommuteTime'])

plt.show()
```

```python
fig2, ax2 = plt.subplots(2, 2)
ax2[0, 1].scatter(x= commute['TripID'], y = commute['CommuteTime'])
ax2[0, 1].scatter(x = [2, 3, 4], y = [12, 13, 14], color = 'red')
ax2[1, 0].scatter(x = commute['CommuteTime'], y = commute['TripID'])

plt.show()
```

```python
fig3, ax3 = plt.subplots(3, 1, figsize = (2,10))
ax3[2].scatter(x = commute['TripID'], y = commute['CommuteTime'])

plt.show()
```

```python
fig4, ax4 = plt.subplots(1, 4, figsize = (8, 1))
ax4[3].scatter(x = commute['TripID'], y = commute['CommuteTime'])

plt.show()
```

```python
fig5, ax5 = plt.subplots()
ax5.bar(height = commute['Date'], x = commute['CommuteTime'])

plt.show()
```

```python
commute.groupby('Vehicle')['CommuteTime'].describe()
```

```python
weather_freq = commute.groupby('Weather')['TripID'].count()
weather_freq.index

fig6, ax6 = plt.subplots()
ax6.bar(x = weather_freq.index, height = weather_freq)
# fig6.title('Title')
plt.show()
```

```python
fig7, ax7 = plt.subplots()
ax7.pie(weather_freq, labels = weather_freq.index)

plt.show()
```
